# if-else statements

```python
# Basic if-else
expression_level = 1500
threshold = 1000

if expression_level > threshold:
    print("Gene is highly expressed")
elif expression_level > 500:
    print("Gene has moderate expression")
else:
    print("Gene has low expression")

# Multiple conditions (important for bioinformatics filters)
read_depth = 30
quality_score = 20
is_variant = True

if read_depth >= 20 and quality_score >= 15 and is_variant:
    print("Variant passes quality filters")
else:
    print("Variant fails quality filters")

# Using 'in' for sequence checking (very useful in bioinformatics!)
codon = "ATG"

if codon in ["ATG", "GTG", "TTG"]:
    print(f"{codon} is a start codon")
elif codon in ["TAA", "TAG", "TGA"]:
    print(f"{codon} is a stop codon")
else:
    print(f"{codon} is a regular codon")
```

# Loops

```python
# For loop - iterate through sequences
protein_sequence = "MVLSPADKTNVKAAWGKVGAHAGEYGAEALERMFLSFPTTKTYFPHF"

print("Amino acid positions:")
for i, aa in enumerate(protein_sequence):
    print(f"Position {i+1}: {aa}")

# For loop with range
print("\nFirst 10 amino acids:")
for i in range(10):
    print(f"{i+1}: {protein_sequence[i]}")

# While loop - useful for parsing until conditions met
fasta_sequences = ["ATCG", "GCTA", "TTAA", "CCGG", "END"]
index = 0

print("\nProcessing FASTA sequences:")
while fasta_sequences[index] != "END":
    print(f"Processing: {fasta_sequences[index]}")
    index += 1

# Loop control with break and continue
print("\nFinding first methionine (M) in protein:")
for i, aa in enumerate(protein_sequence):
    if aa == "M":
        print(f"Found methionine at position {i+1}")
        break  # Exit loop early

# List comprehension (Pythonic way to create lists)
dna_sequence = "ATGCGATACGCTTGA"
rna_sequence = [base.replace('T', 'U') for base in dna_sequence]
print(f"\nDNA: {dna_sequence}")
print(f"RNA: {''.join(rna_sequence)}")
```

# Functions

```python
# Basic function
def calculate_gc_content(sequence):
    """Calculate GC content of a DNA sequence.""" #3 double quotes can write long multiline comments, common to use """ next to definition to tell what this function is about
    sequence = sequence.upper()  # Convert to uppercase
    gc_count = sequence.count('G') + sequence.count('C')
    total_bases = len(sequence)
    
    if total_bases == 0: #== is judgement, = assigning value
        return 0
    
    gc_percentage = (gc_count / total_bases) * 100
    return gc_percentage

# Test the function
test_sequence = "ATGCGATACGCTTGA"
gc_result = calculate_gc_content(test_sequence)
print(f"GC content of '{test_sequence}': {gc_result:.2f}%")

# Function with default parameters
def transcribe_dna_to_rna(dna_sequence, reverse_complement=False):
    """Transcribe DNA to RNA, optionally from reverse complement."""
    if reverse_complement:
        # Create reverse complement
        complement = {'A': 'U', 'T': 'A', 'G': 'C', 'C': 'G'}
        rna = ''.join(complement.get(base, 'N') for base in reversed(dna_sequence))
    else:
        # Simple transcription (T -> U)
        rna = dna_sequence.replace('T', 'U')
    
    return rna

# Test transcription functions
dna_seq = "ATGCGATACGCTTGA"
print(f"DNA: {dna_seq}")
print(f"RNA: {transcribe_dna_to_rna(dna_seq)}")
print(f"RNA from reverse complement: {transcribe_dna_to_rna(dna_seq, reverse_complement=True)}")

# Function returning multiple values
def analyze_sequence(sequence):
    """Analyze a DNA sequence and return multiple statistics."""
    length = len(sequence)
    gc_content = calculate_gc_content(sequence)
    a_count = sequence.upper().count('A')
    t_count = sequence.upper().count('T')
    g_count = sequence.upper().count('G')
    c_count = sequence.upper().count('C')
    
    return length, gc_content, a_count, t_count, g_count, c_count

# Unpack multiple return values
seq_stats = analyze_sequence("ATGCGATACGCTTGA")
print(f"\nSequence statistics: Length={seq_stats[0]}, GC={seq_stats[1]:.2f}%")
```

# Class

```python
# Basic class for a Gene
class Gene:
    """A class to represent a gene."""
    
    def __init__(self, name, sequence, chromosome, start_position, end_position):
        """Initialize a gene with its properties."""
        self.name = name
        self.sequence = sequence.upper()  # Store as uppercase
        self.chromosome = chromosome
        self.start_position = start_position
        self.end_position = end_position
        self.length = len(sequence)
    
    def calculate_gc_content(self):
        """Calculate GC content of the gene."""
        gc_count = self.sequence.count('G') + self.sequence.count('C')
        return (gc_count / self.length) * 100 if self.length > 0 else 0
    
    def transcribe(self):
        """Transcribe DNA to RNA."""
        return self.sequence.replace('T', 'U')
    
    def get_coding_region(self, start_codon="ATG"):
        """Find coding region starting from start codon."""
        start_idx = self.sequence.find(start_codon)
        if start_idx != -1:
            # Return from start codon to end
            return self.sequence[start_idx:]
        return ""
    
    def __str__(self):
        """String representation of the gene."""
        return f"Gene {self.name} on chr{self.chromosome}: {self.start_position}-{self.end_position}"
    
    def __len__(self):
        """Return length of gene."""
        return self.length

# Create a gene object
tp53_gene = Gene(
    name="TP53",
    sequence="ATGCGATACGCTTGAAGCTAGCTAGCT",
    chromosome=17,
    start_position=7668402,
    end_position=7687550
)

# Use the gene object
print(tp53_gene)
print(f"Gene length: {len(tp53_gene)} bp")
print(f"GC content: {tp53_gene.calculate_gc_content():.2f}%")
print(f"RNA transcript: {tp53_gene.transcribe()[:20]}...")
print(f"Coding region starts with: {tp53_gene.get_coding_region()[:10]}")

# More complex class for a Protein
class Protein:
    """A class to represent a protein."""
    
    def __init__(self, name, amino_acid_sequence):
        self.name = name
        self.sequence = amino_acid_sequence
        self.length = len(amino_acid_sequence)
        self.molecular_weight = self.calculate_molecular_weight()
    
    def calculate_molecular_weight(self):
        """Calculate approximate molecular weight in Daltons."""
        # Simple approximation: average amino acid weight = 110 Da
        return self.length * 110
    
    def find_amino_acid_position(self, aa):
        """Find all positions of a specific amino acid."""
        positions = []
        for i, residue in enumerate(self.sequence):
            if residue == aa:
                positions.append(i + 1)  # 1-based indexing
        return positions
    
    def has_motif(self, motif):
        """Check if protein contains a specific motif."""
        return motif in self.sequence
    
    def __str__(self):
        return f"Protein {self.name}: {self.length} aa, ~{self.molecular_weight:.0f} Da"

# Create a protein object
p53_protein = Protein("p53", "MVLSPADKTNVKAAWGKVGAHAGEYGAEALERMFLSFPTTK")
print(f"\n{p53_protein}")
print(f"Lysine (K) positions: {p53_protein.find_amino_acid_position('K')}")
print(f"Contains 'SPAD' motif: {p53_protein.has_motif('SPAD')}")
```
